```python
import pandas as pd

fea_dict = {
   'PSFC': {'str': 'SP', 'unit': 'kPa','convert':{'multiplier':1/1000},'range':[95,100],
             'func':[plot_single_ds], 'func_args':[{}]},
    'WS10': {'str': 'WS', 'unit': 'm/s','func':[plot_single_ds,draw_wind_quiver,plot_maxima], 
             'func_args':[None,{"density":15,"scale":100},{"color":"purple","pt_size":23,"circle":False,"alpha":1,"threshold":0, "min":False}],'vector':["U10","V10"]},
    'QVAPOR_1': {'str': 'QV_1', 'unit': 'kg/kg', 'index':{'bottom_top':1}, 
               'func':[plot_single_ds,draw_contour], 'func_args':[{},{"interval":0.003}]},
    'QVAPOR_6': {'str': 'QV_6', 'unit': 'kg/kg', 'index':{'bottom_top':6}, 
               'func':[plot_single_ds,draw_contour], 'func_args':[{},{"interval":0.003}]},
    'QVAPOR_12': {'str': 'QV_12', 'unit': 'kg/kg', 'index':{'bottom_top':12}, 
               'func':[plot_single_ds,draw_contour], 'func_args':[{},{"interval":0.003}]},
    'QCLOUD_6': {'str': 'QC_6', 'unit': '$kg^{-1}kg$','func':[plot_single_ds], 'func_args':[None], "index":{"bottom_top":6}}, 
    'T2': {'str': 'T2', 'unit': 'K','func':[plot_single_ds], 'func_args':[None]},
}

def calculate_stats(ds, fea_dict):
    stats = {}
    for feature_name, var_dict in fea_dict.items():
        if feature_name in ds:
            ds_var = ds[feature_name]
        else:
            base_feature_name = feature_name.split('_')[0]
            matching_vars = [var for var in ds.variables if var.startswith(base_feature_name)]
            if matching_vars:
                ds_var = ds[matching_vars[0]]
            else:
                raise ValueError(f"Could not find {feature_name} or a similar variable in the dataset")
        
        if 'index' in var_dict:
            ds_var = ds_var.isel(**var_dict['index'])
        
        if 'convert' in var_dict:
            convert = var_dict['convert']
            ds_var = ds_var.copy()
            if 'multiplier' in convert:
                ds_var *= convert['multiplier'] 
            if 'offset' in convert:
                ds_var += convert['offset']
        
        stats[feature_name] = {
            'mean': float(ds_var.mean()),
            'max': float(ds_var.max()),
            'min': float(ds_var.min()),
            'std': float(ds_var.std())
        }
    return stats

narr_stats = calculate_stats(ds_narr, fea_dict)
nam_stats = calculate_stats(ds_nam, fea_dict)

data = []
for var, info in fea_dict.items():
    var_name = info['str']
    var_unit = info['unit']
    
    narr_row = [var, var_name, var_unit, "NARR", narr_stats[var]['mean'], narr_stats[var]['max'], narr_stats[var]['min'], narr_stats[var]['std']]
    nam_row = [var, var_name, var_unit, "NAM", nam_stats[var]['mean'], nam_stats[var]['max'], nam_stats[var]['min'], nam_stats[var]['std']]
    
    if 'index' in info:
        narr_row[0] += f" (index: {info['index']})"
        nam_row[0] += f" (index: {info['index']})"
    
    if 'convert' in info:
        narr_row[0] += f" (convert: {info['convert']})"
        nam_row[0] += f" (convert: {info['convert']})"
    
    data.extend([narr_row, nam_row])

df = pd.DataFrame(data, columns=["Variable", "Name", "Unit", "Dataset", "Mean", "Max", "Min", "Std"])
df[["Name", "Unit", "Dataset", "Mean", "Max", "Min", "Std"]]

# Format the numeric columns to display 2 decimal places using scientific notation
df[["Mean", "Max", "Min", "Std"]] = df[["Mean", "Max", "Min", "Std"]].applymap(lambda x: f"{x:.2e}")

# Display the formatted DataFrame
df[["Name", "Unit", "Dataset", "Mean", "Max", "Min", "Std"]]
```
